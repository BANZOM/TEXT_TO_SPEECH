```python
# importing the module

import pyttsx3 as pts
```

```python
# creating a object of pyttsx3
tools = pts.init()

#getting the voices:
existing_voices=0

for voice in tools.getProperty('voices'):
    print(voice)
    existing_voices= existing_voices + 1

```

```python
print("Total Number of voices are: ", existing_voices)
```

```python
print("You can choose any voice from index \'0\' to index \'%d\' "%(existing_voices-1) ) 
```

```python
voice_sound= int(input("Choose any index from above range: "))
```

```python
# getting the tools ready

voices = tools.getProperty('voices')

tools.setProperty('voice' , voices[voice_sound].id )
```

```python
# let the machine speak

def speak(text):
    tools.save_to_file(text, filename='audioFile.mp3')
    # tools.say(text)
    # tools.runAndWait()

text = input("Enter text to speak: \n")

speak(text)
```
